# Convolution de deux signaux numériques

```python
import numpy as np
import matplotlib.pyplot as plt

# Fonction pour générer le signal unitaire (u[n])
def unit_step(n, shift=0):
    return (n >= shift).astype(int)

# Fonction pour générer une impulsion unitaire (δ[n])
def unit_impulse(n, shift=0):
    return (n == shift).astype(int)

# Fonction pour générer un Peigne de Dirac (p[n])
def dirac_comb(n, period):
    comb = np.zeros_like(n, dtype=int)
    comb[n % period == 0] = 1  # Impulsion à chaque multiple de la période
    return comb

# Fonction pour interpréter une équation entrée par l'utilisateur
def parse_equation(n):
    print("\nEntrez l'équation sous forme interactive.")
    print("Exemple : (2*δ[n] + 5*δ[n-5]) * (u[n+1] - u[n-1])")
    
    x = np.zeros_like(n, dtype=float)
    
    while True:
        print("\nAjoutez un composant au signal :")
        print("1. Impulsion unitaire δ[n]")
        print("2. Impulsion unitaire décalée δ[n-k]")
        print("3. Étape unitaire u[n]")
        print("4. Étape unitaire décalée u[n-k]")
        print("5. Peigne de Dirac p[n]")
        print("6. Terminer et générer le signal")
        
        try:
            choice = int(input("Votre choix : "))
        except ValueError:
            print("Veuillez entrer un numéro valide.")
            continue
        
        if choice == 1:  # Impulsion unitaire δ[n]
            coeff = float(input("Entrez le coefficient de δ[n] : "))
            x += coeff * unit_impulse(n)
        
        elif choice == 2:  # Impulsion unitaire décalée δ[n-k]
            k = int(input("Entrez la valeur de décalage k pour δ[n-k] : "))
            coeff = float(input(f"Entrez le coefficient de δ[n-{k}] : "))
            x += coeff * unit_impulse(n, shift=k)
        
        elif choice == 3:  # Étape unitaire u[n]
            coeff = float(input("Entrez le coefficient de u[n] : "))
            x += coeff * unit_step(n)
        
        elif choice == 4:  # Étape unitaire décalée u[n-k]
            k = int(input("Entrez la valeur de décalage k pour u[n-k] : "))
            coeff = float(input(f"Entrez le coefficient de u[n-{k}] : "))
            x += coeff * unit_step(n, shift=k)
        
        elif choice == 5:  # Peigne de Dirac p[n]
            period = int(input("Entrez la période T du Peigne de Dirac p[n] : "))
            coeff = float(input("Entrez le coefficient du Peigne de Dirac : "))
            x += coeff * dirac_comb(n, period)
        
        elif choice == 6:  # Terminer
            break
        
        else:
            print("Choix invalide. Réessayez.")
    
    return x

# Fonction pour calculer et afficher l'équation de convolution
def compute_convolution_equation(y, n_y):
    equation_terms = []
    for coef, n_val in zip(y, n_y):
        if coef != 0:  # Ignorer les termes avec coefficient nul
            if n_val == 0:
                term = f"{int(coef)}δ[n]"
            elif n_val > 0:
                term = f"{int(coef)}δ[n-{n_val}]"
            else:  # n_val < 0
                term = f"{int(coef)}δ[n+{-n_val}]"
            equation_terms.append(term)
    
    equation = " + ".join(equation_terms)
    return f"y[n] = {equation}"

# Fonction pour tracer et donner l'équation
def plot_signal_and_convolution_with_equation(x, h, n_x, n_h):
    # Convolution des deux signaux
    y = np.convolve(x, h, mode='full') # Le mode 'full' calcule la convolution complète, incluant toutes les superpositions possibles entre x et h.
    n_y = np.arange(n_x[0] + n_h[0], n_x[-1] + n_h[-1] + 1)

    # Tracer les signaux
    plt.figure(figsize=(12, 8))

    # Signal x[n]
    plt.subplot(3, 1, 1)
    plt.stem(n_x, x, basefmt=" ")
    plt.title('$x1[n]$ (1er Signal d\'entrée)')
    plt.xlabel('n')
    plt.ylabel('Amplitude')
    plt.grid()

    # Signal h[n]
    plt.subplot(3, 1, 2)
    plt.stem(n_h, h, basefmt=" ")
    plt.title('$x2[n]$ (2ème Signal d\'entrée)')
    plt.xlabel('n')
    plt.ylabel('Amplitude')
    plt.grid()

    # Convolution y[n]
    plt.subplot(3, 1, 3)
    plt.stem(n_y, y, basefmt=" ")
    plt.title('Convolution $y[n] = x1[n] * x2[n]$')
    plt.xlabel('n')
    plt.ylabel('Amplitude')
    plt.grid()

    plt.tight_layout()
    plt.show()

    # Calcul et affichage de l'équation mathématique
    equation = compute_convolution_equation(y, n_y)
    print("\nÉquation mathématique de la convolution :")
    print(equation)

# Définir la plage de n
n = np.arange(-10, 21)

# Entrer les signaux x1[n] et x2[n]
print("Définir le signal x1[n] :")
x1 = parse_equation(n)
print("\nDéfinir le signal x2[n] :")
x2 = parse_equation(n)

# Tracer les signaux et afficher l'équation de convolution
plot_signal_and_convolution_with_equation(x1, x2, n, n)

```

# Spectrum

```python
import numpy as np
import matplotlib.pyplot as plt

# Fonction pour générer un signal analogique x(t) à partir d'une expression entrée par l'utilisateur
def generate_analog_signal_from_input(t):
    print("\nEntrez votre fonction pour le signal analogique x(t).")
    print("Exemple : 5 * np.cos(2 * np.pi * 2000 * t) + 3 * np.cos(2 * np.pi * 3000 * t)")
    func_str = input("x(t) = ")
    try:
        # Évaluer la fonction entrée
        signal = eval(func_str) # 'eval' interprète la chaîne 'func_str' comme une expression Python et la calcule.
        return signal, func_str
    except Exception as e:
        print(f"Erreur dans l'évaluation de la fonction : {e}")
        return None, None  #Premier None pour le signal et Deuxième None pour func_str

# Fonction pour afficher le signal d'origine
def plot_original_signal(t, signal, title):
    plt.figure(figsize=(12, 6))
    plt.plot(t, signal, label="Signal d'origine", linewidth=1.5, color='blue')
    plt.title(title)
    plt.xlabel("Temps (s)")
    plt.ylabel("Amplitude")
    plt.grid()
    plt.legend()
    plt.show()

# Fonction pour calculer et tracer le spectre unilatéral
def plot_unilateral_spectrum(signal, sampling_rate, title, min_freq, max_freq):
    fft_result = np.fft.fft(signal)  #La Fast Fourier Transform (FFT) est utilisée pour transformer le signal du domaine temporel vers le domaine fréquentiel.
    fft_magnitude = np.abs(fft_result[:len(fft_result)//2]) / len(signal)  #La magnitude (module) des valeurs complexes est calculée à l'aide de np.abs.
                                                                            #Seules les premières moitiés des fréquences (jusqu'à len(fft_result)//2) sont conservées pour obtenir le spectre unilatéral (fréquences positives).
                                                                            #Une normalisation est appliquée en divisant par la longueur du signal pour ajuster les amplitudes.
    fft_magnitude[1:] *= 2  # Multiplication par 2 sauf pour la DC (fréquence 0). 
                             #Cela est nécessaire car, lors du passage au spectre unilatéral, l'énergie des fréquences positives et négatives est combinée
                            #fft_magnitude :C'est un tableau contenant les amplitudes correspondantes aux fréquences calculées.

    freq = np.fft.fftfreq(len(signal), d=1 / sampling_rate)[:len(fft_result)//2] #d=1/sampling_rate représente l'intervalle temporel entre deux échantillons

    plt.figure(figsize=(12, 6))
    plt.stem(freq, fft_magnitude, basefmt=" ", label="Spectre unilatéral", linefmt='b-', markerfmt='bo')
    plt.title(f"Spectre unilatéral : {title}")
    plt.xlabel("Fréquence (Hz)")
    plt.ylabel("Amplitude")
    plt.xlim(min_freq, max_freq)
    for f, a in zip(freq, fft_magnitude):
        if a > 0.01:  # Afficher les fréquences significatives uniquement
            plt.text(f, a, f"{f:.0f} Hz", fontsize=9, color='red', ha='center')
    plt.grid(True)
    plt.legend(loc='upper right')
    plt.show()

# Fonction pour calculer et tracer le spectre bilatéral
def plot_bilateral_spectrum(signal, sampling_rate, title, min_freq, max_freq):
    fft_result = np.fft.fft(signal)
    fft_magnitude = np.abs(fft_result) / len(signal)
    freq = np.fft.fftfreq(len(signal), d=1 / sampling_rate)

    plt.figure(figsize=(12, 6))
    plt.stem(freq, fft_magnitude, basefmt=" ", label="Spectre bilatéral", linefmt='b-', markerfmt='bo')
    plt.title(f"Spectre bilatéral : {title}")
    plt.xlabel("Fréquence (Hz)")
    plt.ylabel("Amplitude")
    plt.xlim(min_freq, max_freq)
    plt.grid(True)
    plt.legend(loc='upper right')
    plt.show()

# Fonction pour calculer et tracer le spectre échantillonné avec ses répliques
def plot_sampled_spectrum(signal, sampling_rate, title, min_freq, max_freq):
    fft_result = np.fft.fft(signal)
    fft_magnitude = np.abs(fft_result) / len(signal)  # Normalisation
    freq = np.fft.fftfreq(len(signal), d=1 / sampling_rate)

    # Calculer le nombre de répliques nécessaires pour couvrir toute la plage de fréquences spécifiée par l'utilisateur
    min_replica = int(np.floor(min_freq / sampling_rate))  # Première réplique
    max_replica = int(np.ceil(max_freq / sampling_rate))   # Dernière réplique

    plt.figure(figsize=(12, 6))

    # Répliques périodiques
    for n in range(min_replica, max_replica + 1):
        shifted_freq = freq + n * sampling_rate
        plt.stem(shifted_freq, fft_magnitude, basefmt=" ", linefmt='b-', markerfmt='bo', label="Répliques" if n == 0 else "")

    plt.title(f"Spectre échantillonné avec répliques : {title}")
    plt.xlabel("Fréquence (Hz)")
    plt.ylabel("Amplitude")
    plt.xlim(min_freq, max_freq)  # Assurer que la plage de fréquence est correctement ajustée
    plt.grid(True)
    plt.legend(loc='upper right')
    plt.show()

# Fonction principale
def main():
    sampling_rate = 1e6  # Utilisation d'un grand taux d'échantillonnage pour représenter les signaux continus
                        #sampling_rate fait généralement référence à la fréquence d'échantillonnage du signal original ou du signal qui a été enregistré ou mesuré.
                        #C'est la fréquence à laquelle les échantillons du signal sont pris pour la conversion du signal analogique en signal numérique.
    duration = 0.01  # Durée du signal (10 ms)
    t = np.arange(0, duration, 1 / sampling_rate)  # Axe temporel

    signal, func_str = None, None
    while signal is None:
        signal, func_str = generate_analog_signal_from_input(t)

    min_freq = float(input("Entrez la fréquence minimale à afficher (en Hz) : "))
    max_freq = float(input("Entrez la fréquence maximale à afficher (en Hz) : "))

    # Tracer le signal d'origine
    plot_original_signal(t, signal, "Signal d'origine")

    # Spectres avant échantillonnage
    plot_unilateral_spectrum(signal, sampling_rate, "Signal analogique", min_freq, max_freq)
    plot_bilateral_spectrum(signal, sampling_rate, "Signal analogique", min_freq, max_freq)

   # Échantillonnage
    Fs = float(input("Entrez la fréquence d'échantillonnage (en Hz) : "))
    sampled_signal = signal[::int(sampling_rate / Fs)]
    sampled_t = t[::int(sampling_rate / Fs)]
    plot_sampled_spectrum(sampled_signal, Fs, "Signal échantillonné", min_freq, max_freq)

if __name__ == "__main__":
    main()
```

# Filter( Filtre numérique idéal)

```python
import numpy as np
import matplotlib.pyplot as plt

# Fonction pour appliquer un filtre sur le spectre
def apply_frequency_filter(fft_result, freq, filter_type, cutoff_freqs, sampling_rate):
    mask = np.zeros_like(freq, dtype=bool) #Cette ligne crée un tableau mask de la même taille que le tableau freq, initialisé avec des valeurs False (0). 
                                            #Ce masque sera utilisé pour sélectionner les fréquences qui doivent être conservées ou filtrées dans le domaine fréquentiel.
    
    # Vérification de la fréquence de coupure pour les fréquences trop élevées
    max_freq = sampling_rate / 2  # Fréquence Nyquist: elle détermine la plus haute fréquence qui peut être correctement 
                                     #échantillonnée sans qu'il n'y ait de phénomène de repliement spectral (ou aliasing).
    for cutoff in cutoff_freqs:
        if cutoff > max_freq:
            print(f"Attention : la fréquence de coupure {cutoff} Hz dépasse la fréquence de Nyquist {max_freq} Hz. Elle sera ajustée.")
    cutoff_freqs = [min(cutoff, max_freq) for cutoff in cutoff_freqs]

    # Application du filtre en fonction du type de filtre
    if filter_type == 'LPF':  # Passe-bas
        mask[np.abs(freq) <= cutoff_freqs[0]] = True
    elif filter_type == 'HPF':  # Passe-haut
        mask[np.abs(freq) >= cutoff_freqs[0]] = True
    elif filter_type == 'BPF':  # Passe-bande
        mask[(np.abs(freq) >= cutoff_freqs[0]) & (np.abs(freq) <= cutoff_freqs[1])] = True
    elif filter_type == 'BSF':  # Coupe-bande
        mask[(np.abs(freq) < cutoff_freqs[0]) | (np.abs(freq) > cutoff_freqs[1])] = True
    else:
        raise ValueError("Type de filtre invalide.")

    # Filtrer le spectre en appliquant le masque
    filtered_spectrum = fft_result * mask

    # Appliquer un seuil pour garantir que les impulsions non pertinentes sont filtrées (ex: LPF > cutoff_freqs)
    if filter_type == 'LPF':
        filtered_spectrum[np.abs(freq) >= cutoff_freqs[0]] = 0  # Supprimer les fréquences supérieures à la coupure

    return filtered_spectrum, mask

# Fonction pour afficher le spectre filtré
def plot_filtered_spectrum(filtered_spectrum, freq, title, min_freq, max_freq):
    fft_magnitude = np.abs(filtered_spectrum) / len(filtered_spectrum)  # Normalisation: il est souvent important de normaliser ces valeurs pour éviter
                                                                        #que les valeurs dans le domaine de fréquence ne soient trop grandes en fonction de la longueur du signal.

    # Affichage du spectre filtré
    plt.figure(figsize=(10, 5))
    plt.stem(freq, fft_magnitude, basefmt=" ", linefmt='r-', markerfmt='ro', label="Spectre filtré")
    plt.title(title)
    plt.xlabel("Fréquence (Hz)")
    plt.ylabel("Amplitude")
    plt.grid()
    plt.legend()
    plt.xlim(min_freq, max_freq)
    plt.show()

# Fonction principale
def main():
    # Exemple de signal et spectre
    sampling_rate = 1e6  # Fréquence d'échantillonnage (1 MHz)
    duration = 0.01  # Durée du signal (10 ms)
    t = np.arange(0, duration, 1 / sampling_rate)  # Axe temporel

    # Exemple de signal (modifiez-le selon vos besoins)
    signal = 1 + 1 * np.cos(2 * np.pi * 2000 * t) + 2 * np.cos(2 * np.pi * 4000 * t) + 3 * np.cos(2 * np.pi * 6000 * t)

    # Spectre du signal
    fft_result = np.fft.fft(signal)
    freq = np.fft.fftfreq(len(signal), d=1 / sampling_rate)

    # Affichage du spectre avant filtrage
    plt.figure(figsize=(12, 6))
    plt.stem(freq, np.abs(fft_result) / len(fft_result), basefmt=" ", linefmt='b-', markerfmt='bo', label="Spectre bilatéral d'origine")
    plt.title("Spectre bilatéral du signal original")
    plt.xlabel("Fréquence (Hz)")
    plt.ylabel("Amplitude")
    plt.grid(True)
    plt.legend()
    plt.xlim(-10000, 10000)  # Limiter l'affichage des fréquences
    plt.show()

    # Demande à l'utilisateur de choisir le filtre
    filter_type = input("Choisissez le type de filtre (LPF, HPF, BPF, BSF) : ").strip().upper() #.strip(): utilisée pour supprimer les espaces au début et à la fin de la chaîne de caractères entrée par l'utilisateur.
                                                                                                #.upper() : Cette méthode convertit la chaîne de caractères saisie par l'utilisateur en majuscules. Cela permet de rendre 
                                                                                                    #l'entrée de l'utilisateur insensible à la casse
    
    if filter_type in ['LPF', 'HPF']:
        cutoff_freqs = [float(input("Entrez la fréquence de coupure (en Hz) : "))]
    elif filter_type in ['BPF', 'BSF']:
        cutoff_freqs = [
            float(input("Entrez la fréquence de coupure basse (en Hz) : ")),
            float(input("Entrez la fréquence de coupure haute (en Hz) : "))]

    # Appliquer le filtre
    filtered_spectrum, mask = apply_frequency_filter(fft_result, freq, filter_type, cutoff_freqs, sampling_rate)

    # Afficher le spectre filtré
    plot_filtered_spectrum(filtered_spectrum, freq, f"Spectre filtré : {filter_type}", -10000, 10000)

    # Reconstitution du signal à partir du spectre filtré
    reconstructed_signal = np.fft.ifft(filtered_spectrum)

    # Affichage du signal filtré dans le domaine temporel
    plt.figure(figsize=(12, 6))
    plt.plot(t, np.real(reconstructed_signal), label="Signal filtré", linewidth=1.5, color='green')
    plt.title(f"Signal filtré dans le domaine temporel")
    plt.xlabel("Temps (s)")
    plt.ylabel("Amplitude")
    plt.grid()
    plt.legend()
    plt.show()

if __name__ == "__main__":
    main()
```

# Filtres numériques: FIR/ IIR

Les filtres numériques sont bien entendu limités par leur fréquence d’échantillonnage et naturellement inadaptés au traitement du signal analogique. La réponse impulsionnelle d’un filtre numérique est un signal à temps discret échantillonné à une fréquence d’échantillonnage fe Par ailleurs, lorsque la réponse impulsionnelle est à durée limitée, c’est à dire h[n]=0 si n>N alors le filtre est à réponse impulsionnelle finie (RIF). Dans le cas contraire, on parle de filtre à réponse impulsionnelle infinie (RII). Ces deux types de filtres sont présentés ci-après, ainsi que deux méthodes de conception

la méthode des fenêtres pour les filtreFIRRF,

la transformation bilinéaire pour liltres IIRRII..

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import firwin, freqz, lfilter, butter

# Fonction pour concevoir un filtre FIR (Finite Impulse Response)
def design_fir_filter(num_taps, cutoff_freqs, sampling_rate, filter_type='lowpass', window='hamming'):
    """
    Conçoit un filtre FIR avec la méthode de la fenêtre.
    
    Arguments :
    - num_taps : nombre de coefficients du filtre (ordre + 1).
    - cutoff_freqs : fréquences de coupure (en Hz). Peut être une valeur unique ou une liste pour les filtres passe-bande ou coupe-bande.
    - sampling_rate : fréquence d'échantillonnage (en Hz).
    - filter_type : type de filtre ('lowpass', 'highpass', 'bandpass', 'bandstop').
    - window : type de fenêtre à utiliser (ex : 'hamming', 'hann', 'blackman')
               Réduction des ondulations dans le spectre
    La méthode des fenêtres est un compromis pratique pour concevoir des filtres FIR. 
    Elle permet de tronquer la réponse impulsionnelle infinie d’un filtre idéal tout en contrôlant les imperfections introduites dans le spectre fréquentiel.
    
    Retourne :
    - Les coefficients du filtre FIR.
    """
    nyquist = sampling_rate / 2  # Fréquence de Nyquist (la moitié de la fréquence d'échantillonnage)
    normalized_cutoff = np.array(cutoff_freqs) / nyquist  # Normalisation des fréquences de coupure
    
    # Conception du filtre FIR en utilisant la méthode de la fenêtre
    fir_coeffs = firwin(num_taps, normalized_cutoff, pass_zero=filter_type, window=window)
    return fir_coeffs

# Fonction pour afficher la réponse en fréquence du filtre FIR ou IIR
def plot_frequency_response(coeffs, sampling_rate, filter_type):
    """
    Affiche la réponse en fréquence du filtre (FIR ou IIR).
    
    Arguments :
    - coeffs : coefficients du filtre (FIR ou IIR).
    - sampling_rate : fréquence d'échantillonnage (en Hz).
    - filter_type : type de filtre ('lowpass', 'highpass', etc.)
    """
    if len(coeffs) == 2:  # Cas d'un filtre IIR (coeffs = [b, a])
        b, a = coeffs
        freq, response = freqz(b, a, worN=8000, fs=sampling_rate)
        filter_label = "IIR"  # Pour un filtre IIR
    else:  # Cas d'un filtre FIR (coeffs est un tableau de coefficients FIR)
        freq, response = freqz(coeffs, worN=8000, fs=sampling_rate)
        filter_label = "FIR"  # Pour un filtre FIR
    
    plt.figure(figsize=(12, 6))
    plt.plot(freq, 20 * np.log10(np.abs(response)), 'b')
    plt.title(f"Réponse en fréquence du filtre {filter_label}")
    plt.xlabel("Fréquence (Hz)")
    plt.ylabel("Amplitude (dB)")
    plt.grid()
    plt.show()

# Fonction pour appliquer un filtre FIR au signal
def apply_fir_filter(signal, fir_coeffs):
    """
    Applique un filtre FIR à un signal.
    
    Arguments :
    - signal : le signal d'entrée.
    - fir_coeffs : coefficients du filtre FIR.
    
    Retourne :
    - Le signal filtré.
    """
    # Application du filtre FIR avec la fonction lfilter
    filtered_signal = lfilter(fir_coeffs, 1.0, signal)
    return filtered_signal

# Fonction pour appliquer un filtre IIR (Infinite Impulse Response) au signal
def apply_iir_filter(signal, sampling_rate, cutoff_freq, filter_type='lowpass', order=4):
    """
    Applique un filtre IIR à un signal.
    
    Arguments :
    - signal : le signal d'entrée.
    - sampling_rate : fréquence d'échantillonnage (en Hz).
    - cutoff_freq : fréquence de coupure (en Hz).
    - filter_type : type de filtre ('lowpass', 'highpass').
    - order : ordre du filtre (le degré du filtre).
    
    Retourne :
    - Le signal filtré.
    """
    nyquist = sampling_rate / 2  # Fréquence de Nyquist
    normalized_cutoff = cutoff_freq / nyquist  # Normalisation de la fréquence de coupure
    b, a = butter(order, normalized_cutoff, btype=filter_type)  # Conception du filtre IIR
    
    # Application du filtre IIR
    filtered_signal = lfilter(b, a, signal)
    return filtered_signal, b, a

# Fonction pour afficher le spectre du signal (transformée de Fourier)
def plot_signal_spectrum(signal, sampling_rate, title="Spectre d'amplitude du signal"):
    """
    Affiche le spectre d'amplitude du signal.
    
    Arguments :
    - signal : le signal d'entrée.
    - sampling_rate : fréquence d'échantillonnage (en Hz).
    - title : titre du graphique.
    """
    n = len(signal)  # Taille du signal
    freqs = np.fft.fftfreq(n, d=1/sampling_rate)  # Fréquences associées au spectre
    spectrum = np.fft.fft(signal)  # Transformée de Fourier du signal
    
    # Affichage du spectre du signal
    plt.figure(figsize=(12, 6))
    plt.plot(np.fft.fftshift(freqs), np.fft.fftshift(np.abs(spectrum)) / n, 'r')
    plt.title(title)
    plt.xlabel("Fréquence (Hz)")
    plt.ylabel("Amplitude")
    plt.grid()
    plt.show()

# Fonction principale
def main():
    # Paramètres
    sampling_rate =15000  # Fréquence d'échantillonnage (Hz)
    duration = 1  # Durée du signal (s)
    t = np.arange(0, duration, 1 / sampling_rate)  # Axe temporel

    # Signal d'entrée : mélange de fréquences
    signal = (1 * np.cos(2 * np.pi * 2000 * t) + 2 * np.cos(2 * np.pi * 4000 * t) + 3 * np.cos(2 * np.pi * 6000 * t))

    # Affichage du spectre du signal original
    plot_signal_spectrum(signal, sampling_rate, title="Spectre d'amplitude du signal original")

    # Conception du filtre FIR
    num_taps = 101  # Ordre du filtre + 1
    cutoff_freqs = [5000]  # Fréquence de coupure pour le filtre passe-bas
    filter_type = 'lowpass'  # Type de filtre
    fir_coeffs = design_fir_filter(num_taps, cutoff_freqs, sampling_rate, filter_type)

    # Affichage de la réponse en fréquence du filtre FIR
    plot_frequency_response(fir_coeffs, sampling_rate, filter_type)

    # Appliquer le filtre FIR au signal
    filtered_signal_fir = apply_fir_filter(signal, fir_coeffs)

    # Afficher le spectre du signal filtré FIR
    plot_signal_spectrum(filtered_signal_fir, sampling_rate, title="Spectre du signal filtré (FIR)")

    # Application du filtre IIR
    cutoff_freq = 5000  # Fréquence de coupure pour le filtre IIR
    filtered_signal_iir, b, a = apply_iir_filter(signal, sampling_rate, cutoff_freq)

    # Affichage de la réponse en fréquence du filtre IIR
    plot_frequency_response((b, a), sampling_rate, filter_type)

    # Affichage du spectre du signal filtré IIR
    plot_signal_spectrum(filtered_signal_iir, sampling_rate, title="Spectre du signal filtré (IIR)")

    # Afficher les signaux avant et après filtrage (FIR et IIR)
    plt.figure(figsize=(12, 6))
    plt.plot(t, signal, label="Signal original")
    plt.plot(t, filtered_signal_fir, label="Signal filtré (FIR)", linestyle="--")
    plt.plot(t, filtered_signal_iir, label="Signal filtré (IIR)", linestyle="--")
    plt.title("Signal avant et après filtrage")
    plt.xlabel("Temps (s)")
    plt.ylabel("Amplitude")
    plt.legend()
    plt.grid()
    plt.show()

# Lancer le programme principal
if __name__ == "__main__":
    main()
```
